```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Airplane_Crashes_V2.csv')
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
df.set_index('DateTime', inplace=True)
```

```python
# Time Series Analysis
plt.figure(figsize=(14, 5))

counts = df.resample('10Y').size()
counts.plot(marker='o', linestyle='-', color= 'red')


plt.title('Number of Crashes Every 10 Years')
plt.xlabel('Time')
plt.ylabel('Number of Crashes')
plt.legend()

for i, count in enumerate(counts):
    plt.text(counts.index[i], count + 5, str(count), ha='center', va='bottom')

plt.show()
```

```python
# Find the year with the most crashes
most_crashes_year = counts.idxmax().year

# Filtering the DataFrame for the year with maximum airplane crashes
df_most_crashes_year = df[df.index.year == most_crashes_year]

# Time Series Analysis for the specific year
plt.figure(figsize=(14, 5))

df_most_crashes_year.resample('M').size().plot(marker='o', linestyle='-', color='red')

plt.title(f'Number of Crashes Each Month in {most_crashes_year}')
plt.xlabel('Month')
plt.ylabel('Number of Crashes')

plt.show()
```

```python
# Find the month with the most crashes within the most_crashes_year
most_crashes_month = df_most_crashes_year.resample('M').size().idxmax().month

# Filter the DataFrame for that specific month
df_most_crashes_month = df_most_crashes_year[df_most_crashes_year.index.month == most_crashes_month]

# Time Series Analysis for the specific month
plt.figure(figsize=(14, 5))

# Plotting crashes per day in the most_crashes_month
df_most_crashes_month.resample('D').size().plot(marker='o', linestyle='-', color='brown')

plt.title(f'Number of Crashes Each Day in {most_crashes_year}, Month {most_crashes_month}')
plt.xlabel('Day')
plt.ylabel('Number of Crashes')

plt.show()
```

```python
#Total Accidents by Type of Flight (Military, Passenger):
plt.figure(figsize=(10, 6))
sns.countplot(x='Flight Type', data=df)
plt.title('Total Accidents by Type of Flight')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
yearly_fatalities = df.resample('10Y').sum()['Fatalities']
yearly_fatalities.index = yearly_fatalities.index.year.astype(str)
yearly_fatalities.plot(kind='bar')
plt.title('Number of Fatalities by Year')
plt.xlabel('Year')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=0)
plt.show()
```

```python
#Airline Operators with Most Crashes:
plt.figure(figsize=(12, 6))
sns.countplot(x='Operator', data=df, order=df['Operator'].value_counts().index[:5])
plt.xticks(rotation=45)
plt.title('Top 10 Airline Operators with Most Crashes')
plt.show()
```

```python
#Fatalities by Airline Operators:
plt.figure(figsize=(12, 6))
df.groupby('Operator')['Fatalities'].sum().sort_values(ascending=False)[:5].plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Fatalities by Airline Operators')
plt.xlabel('Airline Operator')
plt.ylabel('Number of Fatalities')
plt.show()
```

```python
#Fatalities by Aircraft Manufacturer:
plt.figure(figsize=(12, 6))
df.groupby('Type')['Fatalities'].sum().sort_values(ascending=False)[:5].plot(kind='bar')
plt.xticks(rotation=45)
plt.title('Fatalities by Aircraft Type (Manufacturer)')
plt.xlabel('Aircraft Type')
plt.ylabel('Number of Fatalities')
plt.show()
```

```python
#Survivors and Fatalities:
plt.figure(figsize=(10, 6))
df['Survivors'] = df['Aboard'] - df['Fatalities']
df[['Survivors', 'Fatalities']].sum().plot(kind='bar', color=['green', 'red'])
plt.title('Survivors and Fatalities')
plt.xlabel('Outcome')
plt.ylabel('Number of People')
plt.show()

```

```python
# Fatalities vs. Aboard
plt.figure(figsize=(10, 6))
plt.scatter(df['Aboard'], df['Fatalities'], alpha=0.5)
plt.title('Fatalities vs. Aboard')
plt.xlabel('Number of People Aboard')
plt.ylabel('Number of Fatalities')
plt.show()
```
